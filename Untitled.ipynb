```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('gyroscope_data_firebase.csv')

# Remove any rows that contain non-numeric values or empty fields
# Drop rows with any NaN (empty) values
df_cleaned = df.dropna()

# Ensure that gx, gy, gz values are numeric (convert errors to NaN and then drop them)
df_cleaned[['gx', 'gy', 'gz']] = df_cleaned[['gx', 'gy', 'gz']].apply(pd.to_numeric, errors='coerce')

# Drop any rows where gx, gy, or gz are NaN (i.e., non-numeric or failed to convert)
df_cleaned = df_cleaned.dropna(subset=['gx', 'gy', 'gz'])

# Save the cleaned data back to a CSV file
df_cleaned.to_csv('gyroscope_data_cleaned.csv', index=False)

print(f"Cleaned data saved to 'gyroscope_data_cleaned.csv'. Total records: {len(df_cleaned)}.")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned CSV file
df = pd.read_csv('gyroscope_data_cleaned.csv')

# Plot gx, gy, gz separately
plt.figure()

# Plot gx
plt.subplot(3, 1, 1)
plt.plot(df['timestamp'], df['gx'], label='gx')
plt.xlabel('Time')
plt.ylabel('gx')
plt.title('Gyroscope gx')

# Plot gy
plt.subplot(3, 1, 2)
plt.plot(df['timestamp'], df['gy'], label='gy', color='orange')
plt.xlabel('Time')
plt.ylabel('gy')
plt.title('Gyroscope gy')

# Plot gz
plt.subplot(3, 1, 3)
plt.plot(df['timestamp'], df['gz'], label='gz', color='green')
plt.xlabel('Time')
plt.ylabel('gz')
plt.title('Gyroscope gz')

# Display the plots
plt.tight_layout()
plt.show()

# Plot all three variables in a single graph
plt.figure()
plt.plot(df['timestamp'], df['gx'], label='gx', color='blue')
plt.plot(df['timestamp'], df['gy'], label='gy', color='orange')
plt.plot(df['timestamp'], df['gz'], label='gz', color='green')

plt.xlabel('Time')
plt.ylabel('Gyroscope values')
plt.title('Gyroscope gx, gy, gz Combined')
plt.legend()
plt.show()
```
